# Cleaning columns title flaws

# Data Analysis and Understanding

<br><b> More than 20% of data missing in these cols </b>

```python
d = dict(df.isnull().sum()[1:])
more_missing = []
for i in d:
    if d[i] > 530:
        more_missing.append(i)
        print(i," - ",d[i], " or ",round((d[i]/df.shape[0])*100),"%")
```

<br><b>Further Analysis of the more_missing columns<br>Data Only Present in Years</b>

```python
print("Currency Exchange Rate data present only of Years")

for j in more_missing:
    l1 = []
    for i in df[df[j].isnull()==False][['Date',j,'(USD)']]['Date']:
        if i.split('-')[-1] not in l1:
            l1.append(i.split('-')[-1])
    print(j," --> ",l1)
```

<br><b> Less than 20% of data missing in these cols</b>

```python
d = dict(df.isnull().sum()[1:])
less_missing = []
for i in d:
    if d[i] < 530 and d[i] != 0:
        less_missing.append(i)
        print(i," - ",d[i], " or ",round((d[i]/df.shape[0])*100),"%")
```

<br><b>Further Analysis of the less_missing columns<br>Data Only Present in Years</b>

```python
print("Currency Exchange Rate data present only of Years")

for j in less_missing:
    l1 = []
    for i in df[df[j].isnull()==False][['Date',j,'(USD)']]['Date']:
        if i.split('-')[-1] not in l1:
            l1.append(i.split('-')[-1])
    print(j," --> ",l1)
```

# Conversion of string type Date into datetime format

```python
df.sort_values(by='Date',ascending=True,inplace=True)
dates = []
for i in df['Date']:
    i = i.replace('-','/')
    change_fmt = "%d/%b/%y"
    dates.append(datetime.strptime(i,change_fmt))
df['Date'] = dates
```

# Data Imputation and Processing

<br><b>Filling data in more_missing columns</b>

```python
df.drop(columns='(VES)',inplace=True)
more_missing.remove('(VES)')
for i in more_missing:
    df[i].fillna("NDA since 2018",inplace=True)
df[df['Date'] < '2018-01-02'] = df[df['Date'] < '2018-01-02'].replace(['NDA since 2018'],method='bfill')
```

<b>Filling data in less_missing columns</b>

```python
for i in less_missing:
    if i != '(COP)':
        df[i].fillna(method='ffill',inplace=True)
    else:
        df['(COP)'].fillna(value='NDA since 2022',inplace=True)
```

# Remaining Missing rows in columns
<b>Finding what kind of row or data is missing in columns</b>

```python
d = dict(df.isnull().sum()!=0)
missing_dates = []
for i in d:
    if d[i]==True:
        missing_dates.append(i)
```

```python
df[df['(GBP)'].isnull()][['Date','(RUB)']]
```

<br><b>Imputing the remaining missing data</b>

```python
for i in missing_dates:
    df[i].fillna(method='bfill',inplace=True)
df.sort_values(by='Date',ascending=True,inplace=True,ignore_index=True)
```

# Final Cleanup

```python
l = []
for i in df.columns:
     l.append(i.split('(')[-1].split(')')[0])
df.columns = l
df.replace('NDA since 2018',0.00,inplace=True)
df.replace('NDA since 2022',0.00,inplace=True)
df.iloc[:,39:].astype('float64')
```

```python
df.sample(5)
```

# Feature Engineering

```python
weeks = []
year = []
months = []
for i in df['Date']:
    weeks.append(datetime.date(i).isocalendar().week)
    year.append(datetime.date(i).isocalendar().year)
    months.append(i.month)
df['Year'] = year
df['Week'] = weeks
df['Month'] = months
```

```python
df.drop(columns='Date',inplace=True)
```

# Storing the processed data

<br><b>Splitting Data Year-Wise</b>

```python
l = []
for i in (2012,2013,2014,2015,2016,2017,2018,2019,2020,2021,2022):
    l.append(df[df['Year']==i])
```

<br><b>Splitting the Year-Wise Data into Month-Wise Yearly Data and Week-Wise Yearly Data</b>

```python
monthly_per_yr = []
for i in l:
    monthly_per_yr.append(i.groupby('Month').mean())
    
week_per_yr = []
for i in l:
    week_per_yr.append(i.groupby('Week').mean())
```

<br><b>Storing the final processed csv files</b>

```python
count = 0
for i,j in zip(monthly_per_yr,week_per_yr):
    i.to_csv(str(2012+count)+'M.csv')
    j.to_csv(str(2012+count)+'W.csv')
    count+=1
```
